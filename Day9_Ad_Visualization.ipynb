```python
import matplotlib.pyplot as plt
```

```python
marks = [35,40,42,45,48,
         50,52,55,58,60,
         62,65.67,68,70,
         72,74,75,78,80,
         82,84,85,88,90,
         92,94,95,97,99]

plt.figure(figsize=(8,5))
plt.hist(marks, bins=7)
plt.title("Distribution of student marks")
plt.xlabel("Marks")
plt.ylabel("Number of students")

plt.grid(True)

plt.show()
```
